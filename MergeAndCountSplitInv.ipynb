# Merge-and-CountSplitInv  
__Input:__ sorted arrays _C_ and _D_ (length _n_/2 each)  
__Output:__ sorted array _B_ (length _n_) and the number of split inversions  
__Simplifying assumption:__ _n_ is even
***
$i:=1, j:=1, splitInv:=0$  
__for__ $k := 1$ to _n_ __do__  
__if__ $C[i] < D[j]$ __then__  
$B[k]:=C[i], i:=i+1$  
__else__ $//D[j]<C[i]$  
$B[k]:=D[j], j:=j+1$  
$splitInv:=splitInv + (n/2 - i + 1)$  
  
$return (B, splitInv)$


```python
#initArray = [1, 3, 5, 2, 4, 6] // returns 3.
initArray = [1,6,3,2,4,5] # should return 4 
mergeAndCountSplitInv(initArray)

def mergeAndCountSplitInv(someArray):
    print("hello")
    B = []
    #C = [1, 3, 5]
    #D = [2, 4, 6]
    C = someArray[:len(someArray)//2]
    D = someArray[len(someArray)//2:]
    print(C, D)
    i = 0
    j = 0
    splitInv = 0
    
    for k in range(0, len(someArray)):
        if C[i] < D[j]:
            print(k , i)
            #B[k] = C[i]
            B.insert(k, C[i])
            i = i + 1
        else:
            #B[k] = D[j]
            B.insert(k, D[j])
            j = j + 1
            splitInv = splitInv + ((len(initArray) / 2) - i)  #refactor to be dynamic to array size

        print(B, splitInv, i, j)


```
